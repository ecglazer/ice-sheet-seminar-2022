# Perfectly plastic ice sheet 

The driving stress is given by

$\tau_d = -\rho_i g H \frac{\partial H}{\partial x}$ 

The simplest ice sheet model we can derive from this is one which assumes a plastic rheology, ie.

$\tau_d = Y = 100$ kPa,

which leads to

$ H = \left(\frac{xY}{2\rho_i g}\right)^\frac{1}{2}$

```python
import numpy as np
import matplotlib.pyplot as plt

Y =100

rho_i = 917

g = 9.8

x = np.array(range(0,1000))

H = np.sqrt(x*Y/2/rho_i/g)


fig, ax = plt.subplots(figsize=(8, 4), layout='constrained')
ax.plot(x, H, label='ice sheet surface') 
ax.set_xlabel('distance, x [m]')
ax.set_ylabel('ice sheet surface height, H [m]');
```
